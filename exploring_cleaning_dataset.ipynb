# ðŸ›’ Cleaning and Exploring Sales Dataset


## Impoting Pandas

```python
import pandas as pd
```

## Load the dataset

```python
df = pd.read_excel('sales.xlsx')


# Inspect the first few rows
df.head()

```

## Check for missing values

```python

print(df.isnull().sum())
```

## Get basic statistics

```python
df.describe()
```

## Fill missing values

```python
df['ADDRESSLINE2'].fillna("N/A", inplace=True)
df['STATE'].fillna("N/A", inplace=True)

print("\nAfter Handling Missing Values:")
df
```

### Convert ORDERDATE to datetime
### Clean PHONE column (remove periods)

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

df['PHONE'] = df['PHONE'].str.replace('.', '')

print("\nAfter Standardizing Data Formats:")
df
```

## Check for duplicates

```python
print("Number of duplicates:", df.duplicated().sum())

# Drop duplicates (if any)
df.drop_duplicates(inplace=True)

print("\nAfter Removing Duplicates:")
df
```

## Check data types

```python
print("\nData Types Before Cleaning:")
print(df.dtypes)

# Convert numeric columns (if needed)
numeric_columns = ['QUANTITYORDERED', 'PRICEEACH', 'SALES']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

print("\nData Types After Cleaning:")
print(df.dtypes)
```

## Clean text columns (remove extra spaces)

```python
text_columns = ['CUSTOMERNAME', 'ADDRESSLINE1', 'CITY', 'STATE', 'COUNTRY']
df[text_columns] = df[text_columns].apply(lambda x: x.str.strip())

print("\nAfter Cleaning Text Columns:")
df
```

#### Drop some bad colums

```python
df.drop(columns=['PHONE','TERRITORY','ADDRESSLINE2'], inplace=True)
df.head()
```

```python
df['POSTALCODE'].fillna(df['POSTALCODE'].mode()[0], inplace=True)
df.head()
```

### Chicking for nulls

```python
df.isnull().sum()
```
