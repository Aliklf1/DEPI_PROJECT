```python
import pandas as pd



# Create DataFrame
df = pd.read_excel("C:\\ymy Data\\Panadas\\my-venv\\DEPI_PROJECT\\cleaned_data.xlsx")


# Normalize into separate tables
orders = df[["ORDERNUMBER", "ORDERDATE", "STATUS", "QTR_ID", "MONTH_ID", "YEAR_ID", "CUSTOMERNAME", "DEALSIZE"]]
order_details = df[["ORDERNUMBER", "PRODUCTCODE", "QUANTITYORDERED", "PRICEEACH", "ORDERLINENUMBER", "SALES"]]
products = df[["PRODUCTCODE", "PRODUCTNAME", "PRODUCTLINE", "MSRP"]]
customers = df[["CUSTOMERNAME", "ADDRESSLINE1", "CITY", "STATE", "POSTALCODE", "COUNTRY", "CONTACTLASTNAME", "CONTACTFIRSTNAME"]]

# Remove duplicates
orders = orders.drop_duplicates()
order_details = order_details.drop_duplicates()
products = products.drop_duplicates()
customers = customers.drop_duplicates()

# Save tables to CSV files
orders.to_csv("orders.csv", index=False)
order_details.to_csv("order_details.csv", index=False)
products.to_csv("products.csv", index=False)
customers.to_csv("customers.csv", index=False)

print("Tables saved successfully without duplicates!")
```
